```python
def get_youtube_transcription(video_url):
    # Extrai o ID do vídeo da URL
    video_id = video_url.split("v=")[-1].split("&")[0]

    try:
        # Obtém a transcrição do YouTube
        transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=['en', 'pt'])
        
        # Junta todas as legendas em um texto contínuo
        lyrics = "\n".join([t["text"] for t in transcript])
        return lyrics

    except Exception as e:
        return f"Erro ao obter transcrição: {str(e)}"
    

# URL do vídeo
video_url = "https://www.youtube.com/watch?v=v3BvJNjQDSI"

# Obtém a transcrição
lyrics = get_youtube_transcription(video_url)
lyrics
```

```python
from youtube_transcript_api import YouTubeTranscriptApi

def get_youtube_transcription(video_url):
    # Extrai o ID do vídeo da URL
    video_id = video_url.split("v=")[-1].split("&")[0]

    try:
        # Obtém a lista de legendas disponíveis
        transcript_list = YouTubeTranscriptApi.list_transcripts(video_id)
        
        # Seleciona a transcrição original (sem tradução)
        transcript = transcript_list.find_generated_transcript(["pt", "en"])

        # Junta todas as legendas em um texto contínuo
        lyrics = "\n".join([t["text"] for t in transcript.fetch()])
        return lyrics

    except Exception as e:
        return f"Erro ao obter transcrição: {str(e)}"

# URL do vídeo
video_url = "https://www.youtube.com/watch?v=v3BvJNjQDSI"

# Obtém a transcrição
lyrics = get_youtube_transcription(video_url)
print(lyrics)
```
